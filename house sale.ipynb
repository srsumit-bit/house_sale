```python
# now i have used heat map to show multiple relationship of this data
plt.figure(figsize=(12, 10))
selected_num = [
    'SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea',
    'TotalBsmtSF', '1stFlrSF', 'YearBuilt', 'FullBath', 'TotRmsAbvGrd', 'Fireplaces'
 ]
sns.heatmap(data[selected_num].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
features_for_heatmap = [
    'SalePrice', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd',
    'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF',
    '1stFlrSF', '2ndFlrSF', 'FullBath', 'HalfBath', 'TotRmsAbvGrd',
    'Fireplaces', 'LotFrontage', 'LotArea', 'MasVnrArea'
]

# Filter the dataset for selected features
data = data[features_for_heatmap]

# Create a heatmap with larger figure size
plt.figure(figsize=(14, 12))
sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Extended Correlation Heatmap of Housing Features", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
